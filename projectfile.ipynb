```python
def transform_data(df, value_columns):
    # Reshaping the DataFrame for each set of value columns
    melted_dfs = []
    for col_prefix in value_columns:
        melted_df = pd.melt(df, id_vars='Company Name', var_name='year', value_name=col_prefix)
        #melted_df['year'] = melted_df['year'].str.extract('(\d+)', expand=False).astype(float)
        #melted_df['year'] = range
        melted_dfs.append(melted_df)
    
    # Merging the melted DataFrames
    result_df = melted_dfs[0]
    for melted_df in melted_dfs[1:]:
        result_df = pd.merge(result_df, melted_df, on=['Company Name'])
    
    # Sorting the DataFrame
    result_df = result_df.sort_values(by=['Company Name']).reset_index(drop=True)
    
    return result_df
```

```python
import pandas as pd

# Create a DataFrame from the provided data
'''
data = {
    'Company Name': ['A C C Ltd.', 'A I A Engineering Ltd.'],
    'TQ2014': [2.130143089, 3.184261722],
    'TQ2015': [2.274329494, 5.879728528],
    'TQ2016': [1.989409051, 3.713890225],
    'TY2014': [1, 4],
    'TY2015': [2, 5],
    'TY2016': [3, 6]
}
'''

#df = pd.DataFrame(data)

# Initialize an empty list to store transformed rows
transformed_rows = []

# Iterate over the rows of the DataFrame
for index, row in df.iterrows():
    company_name = row['Company Name']
    
    # Iterate over the columns of the row
    for col_name, value in row.items():
        if col_name.startswith('TQ') or col_name.startswith('LTA'):
            year = col_name[2:]
            variable = col_name[:2]
            transformed_rows.append([company_name, year, value, variable])

# Create a DataFrame from the transformed rows
transformed_df = pd.DataFrame(transformed_rows, columns=['Company Name', 'year', 'value', 'variable'])

# Pivot the DataFrame to get separate TQ and TY columns
final_df = transformed_df.pivot_table(index=['Company Name', 'year'], columns='variable', values='value').reset_index()

# Display the transformed DataFrame
print(final_df)
```
